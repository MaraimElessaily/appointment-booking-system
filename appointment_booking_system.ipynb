```python
import tkinter as tk
from tkinter import messagebox

# Function to book appointment
def book_appointment():
    name = entry_name.get().title()
    day = entry_day.get().capitalize()
    time = entry_time.get()
    service = entry_service.get().title()

    if not name or not day or not time or not service:
        messagebox.showwarning("Missing Info", "Please fill in all fields!")
        return

    appointment = [name, day, time, service]

    message = (
        "--- APPOINTMENT CONFIRMATION ---\n\n"
        f"Client Name : {appointment[0]}\n"
        f"Day         : {appointment[1]}\n"
        f"Time        : {appointment[2]}\n"
        f"Service     : {appointment[3]}\n"
        f"Total Fields: {len(appointment)}"
    )

    messagebox.showinfo("Booking Confirmed", message)

# Create main window
window = tk.Tk()
window.title("Appointment Booking System")
window.geometry("420x360")
window.configure(bg="#f2f2f2")

# Main Frame (Card Style)
frame = tk.Frame(window, bg="white", padx=20, pady=20)
frame.pack(pady=20)

# Title
title = tk.Label(
    frame,
    text="Appointment Booking System",
    font=("Arial", 16, "bold"),
    bg="white",
    fg="#333"
)
title.grid(row=0, column=0, columnspan=2, pady=(0, 15))

# Labels & Entries
labels = ["Client Name", "Appointment Day", "Appointment Time", "Service Name"]
entries = []

for i, text in enumerate(labels, start=1):
    label = tk.Label(frame, text=text, bg="white", anchor="w")
    label.grid(row=i, column=0, sticky="w", pady=5)

    entry = tk.Entry(frame, width=25)
    entry.grid(row=i, column=1, pady=5)
    entries.append(entry)

entry_name, entry_day, entry_time, entry_service = entries

# Button
btn = tk.Button(
    frame,
    text="Book Appointment",
    command=book_appointment,
    bg="#007acc",
    fg="white",
    padx=10,
    pady=5
)
btn.grid(row=6, column=0, columnspan=2, pady=15)

# Run window
window.mainloop()
```
